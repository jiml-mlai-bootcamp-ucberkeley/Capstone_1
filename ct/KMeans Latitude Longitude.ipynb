```python
from warnings import filterwarnings
filterwarnings("ignore")

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import KMeans
```

```python
zip_lon_lat = pd.read_csv('ct_zip-lon-lat.csv')
zip_lon_lat["zip_code_str"] = np.where(zip_lon_lat.zip_code.apply(int) < 7000,'0'+ zip_lon_lat.zip_code.apply(int).apply(str),zip_lon_lat.zip_code.apply(int).apply(str))
zip_lon_lat.head()
```

```python
features = ["latitude","longitude"]
X = zip_lon_lat[features]
number_of_counties = 8
```

```python
kmeans = KMeans( init="random", n_clusters=number_of_counties,random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(X["longitude"], X["latitude"], c=y_kmeans, s=50, cmap='viridis')
```

```python
zip_lon_lat['county'] = y_kmeans.tolist()
zip_lon_lat.head()
```

```python
zip_lon_lat.to_csv('county.csv')
```

```python
zip_lon_lat.groupby(['county'])['county'].count()
```

```python
outliers_df = pd.read_csv('outliers.csv')
outliers_df.drop(columns=["Unnamed: 0"],inplace=True)
outliers_df["zip_code_str"] = np.where(outliers_df.zip_code.apply(int) < 7000,'0'+ outliers_df.zip_code.apply(int).apply(str),outliers_df.zip_code.apply(int).apply(str))
#outliers_df.head()
```

```python
merged_df = pd.merge(outliers_df, zip_lon_lat, how='inner', on="zip_code_str")
merged_df.tail()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(merged_df["longitude"],merged_df["latitude"], c=merged_df["county"], s=50, cmap='viridis')
```
