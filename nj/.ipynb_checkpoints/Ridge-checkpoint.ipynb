```python
from warnings import filterwarnings
filterwarnings("ignore")

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.feature_selection import SequentialFeatureSelector, RFE
from sklearn.linear_model import Ridge

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
outliers_df = pd.read_csv('outliers-data.csv')
outliers_df.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
outliers_df.info()
```

```python
outliers_df["prev_sold_year"] = pd.to_datetime(outliers_df["prev_sold_date"], format='%Y-%m-%d').dt.year
outliers_df.drop(columns=["prev_sold_date"],inplace=True)
outliers_df.info()
```

```python
num_features_to_select = 5
random_state_value = 42
```

```python
X = outliers_df[["bed","bath","acre_lot","zip_code","house_size","prev_sold_year"]]
y = outliers_df["price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state_value)
```

```python
def print_stuff(title, mse,mae,r2):
    print(title + " MSE = {:,.6f}".format(ridge_mse))
    print(title + " MAE = {:,.6f}".format(ridge_mae))
    print(title + " R2 = {:,.6f}".format(ridge_r2))
```

```python
ridge = Ridge(alpha=1.0)
ridge_model = ridge.fit(X_train,y_train)
ridge_predict = ridge_model.predict(X_test)
ridge_mse = mean_squared_error(ridge_predict,y_test)
ridge_mae = mean_absolute_error(ridge_predict,y_test)
ridge_r2 = r2_score(y_test,ridge_predict)
```

```python
print_stuff("Ridge",ridge_mse,ridge_mae,ridge_r2)
```

```python
ridge_sfs = SequentialFeatureSelector(ridge, n_features_to_select=num_features_to_select)
ridge_sfs.fit(X, y)
ridge_sfs_selected_features = ridge_sfs.get_support()
print('The selected features are:', list(X.columns[ridge_sfs_selected_features]))
```

```python
X_sfs = outliers_df[list(X.columns[ridge_sfs_selected_features])]
y_sfs = outliers_df["price"]
X_sfs_train, X_sfs_test, y_sfs_train, y_sfs_test = train_test_split(X_sfs, y_sfs, test_size=0.3, random_state=random_state_value)
```

```python
ridge_sfs = Ridge()
ridge_sfs_model = ridge.fit(X_sfs_train,y_sfs_train)
ridge_sfs_predict = ridge_sfs_model.predict(X_sfs_test)

ridge_sfs_mse = mean_squared_error(y_sfs_test,ridge_sfs_predict)
ridge_sfs_mae = mean_absolute_error(y_sfs_test,ridge_sfs_predict)
ridge_sfs_r2 = r2_score(y_sfs_test,ridge_sfs_predict)
```

```python
print_stuff("Ridge",ridge_sfs_mse,ridge_sfs_mae,ridge_sfs_r2)
```

```python
ridge_rfe = RFE(ridge, n_features_to_select=num_features_to_select)
ridge_rfe.fit(X, y)
ridge_rfe_selected_features = ridge_rfe.get_support()
print('The selected features are:', list(X.columns[ridge_rfe_selected_features]))
```

```python
X_rfe = outliers_df[list(X.columns[ridge_rfe_selected_features])]
y_rfe = outliers_df["price"]
X_rfe_train, X_rfe_test, y_rfe_train, y_rfe_test = train_test_split(X_rfe, y_rfe, test_size=0.3, random_state=random_state_value)
```

```python
ridge_rfe = Ridge()
ridge_rfe_model = ridge.fit(X_rfe_train,y_rfe_train)
ridge_rfe_predict = ridge_rfe_model.predict(X_rfe_test)

ridge_rfe_mse = mean_squared_error(y_rfe_test,ridge_rfe_predict)
ridge_rfe_mae = mean_absolute_error(y_rfe_test,ridge_rfe_predict)
ridge_rfe_r2 = r2_score(y_rfe_test,ridge_rfe_predict)
```

```python
print_stuff("Ridge",ridge_rfe_mse,ridge_rfe_mae,ridge_rfe_r2)
```

```python
ridge_predict_df = pd.DataFrame(ridge_predict).reset_index()
ridge_predict_df.rename(columns={0: "prediction"},inplace=True)

ridge_sfs_predict_df = pd.DataFrame(ridge_sfs_predict).reset_index()
ridge_sfs_predict_df.rename(columns={0: "prediction"},inplace=True)

ridge_rfe_predict_df = pd.DataFrame(ridge_rfe_predict).reset_index()
ridge_rfe_predict_df.rename(columns={0: "prediction"},inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.ticklabel_format(style='plain')
ax.yaxis.set_major_formatter('${x:1,.0f}')
ax.xaxis.set_major_formatter(' ')

plt.scatter(ridge_predict_df.index,ridge_predict_df["prediction"], color="black")
plt.scatter(ridge_sfs_predict_df.index,ridge_predict_df["prediction"], color="blue")
plt.scatter(ridge_rfe_predict_df.index,ridge_predict_df["prediction"], color="green")
plt.show()
```
